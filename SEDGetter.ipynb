```python
from spitzer_plots import get_spitzer_data
```

```python
crd = SkyCoord(351.75249*u.deg, -0.526919*u.deg, frame='galactic').fk5
```

```python
spitz_files = get_spitzer_data(crd, 1*u.arcmin)
```

```python
%matplotlib inline
import pylab as pl
pl.style.use('dark_background')
```

```python
from astropy import wcs
```

```python
pl.figure(figsize=(16,4))
for ii,band in enumerate(spitz_files):
    ww = wcs.WCS(spitz_files[band][0].header)
    ax = pl.subplot(1,5,ii+1, projection=ww)
    ax.imshow(spitz_files[band][0].data)
```
